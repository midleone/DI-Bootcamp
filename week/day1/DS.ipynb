```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

cities = ['City1', 'City2', 'City3', 'City4', 'City5', 'City6', 'City7', 'City8', 'City9', 'City10']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

temperature_df = pd.DataFrame(temperature_data, index=cities, columns=months)

print(temperature_df)
```

```python
annual_avg_temperature = temperature_df.mean(axis=1)

hottest_city = annual_avg_temperature.idxmax()
coldest_city = annual_avg_temperature.idxmin()

print("Annual Average Temperature:")
print(annual_avg_temperature)

print("\nCity with the Highest Average Temperature:", hottest_city)
print("City with the Lowest Average Temperature:", coldest_city)
```

```python
for city in temperature_df.index:
    if city == hottest_city:
        plt.plot(months, temperature_df.loc[city], label=city, linewidth=2, linestyle='--', color='red')
    elif city == coldest_city:
        plt.plot(months, temperature_df.loc[city], label=city, linewidth=2, linestyle='-.', color='blue')
    else:
        plt.plot(months, temperature_df.loc[city], label=city, linewidth=1, linestyle='-', color='gray')

plt.title('Monthly Temperature Trends for Each City')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.legend()

plt.show()
```
