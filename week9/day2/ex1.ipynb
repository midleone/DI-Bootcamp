```python
import pandas as pd

file_path = 'data.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
data.info()
```

```python
# Attempting to read the file with a different encoding
try:
    data = pd.read_csv(file_path, encoding='ISO-8859-1')
except Exception as e:
    error = e

# Check if the file is loaded successfully or return the error
if 'data' in locals():
    successfully_loaded = True
    data_info = data.info()
    first_rows = data.head()
else:
    successfully_loaded = False
    data_info, first_rows = error, None

successfully_loaded, data_info, first_rows
```

```python
# Converting 'InvoiceDate' to datetime format
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Finding the most recent date in the dataset
most_recent_date = data['InvoiceDate'].max()

# Calculating Recency for each customer
# Note: Assuming CustomerID is a unique identifier for each customer
recency_data = data.groupby('CustomerID').agg(
    Last_Purchase_Date=('InvoiceDate', 'max')
)
recency_data['Recency'] = (most_recent_date - recency_data['Last_Purchase_Date']).dt.days

# Display the first few rows of the recency data
recency_data.head(), most_recent_date
```

```python
# Calculating Frequency for each customer
frequency_data = data.groupby('CustomerID').agg(
    Frequency=('InvoiceNo', 'nunique')
)

# Display the first few rows of the frequency data
frequency_data.head()
```
