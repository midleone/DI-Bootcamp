```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = {'x': np.random.rand(20), 'y': np.random.rand(20)}
df = pd.DataFrame(data)

sns.scatterplot(x='x', y='y', data=df)

plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('Basic Scatter Plot')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
categories = np.random.choice(['A', 'B', 'C'], size=30)
df = pd.DataFrame({'category': categories})

category_counts = df['category'].value_counts()
plt.bar(category_counts.index, category_counts.values)

plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Category Frequency')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(0, 10, 100)
df = pd.DataFrame({'x': x_values, 'y1': np.sin(x_values), 'y2': np.cos(x_values)})

sns.lineplot(x='x', y='y1', data=df, label='sin(x)')
sns.lineplot(x='x', y='y2', data=df, label='cos(x)')

plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('Comparison of Two Functions')

plt.legend()

plt.show()
```
