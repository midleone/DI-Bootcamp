```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv'
df = pd.read_csv(url)
```

```python
for col in df.select_dtypes(include='number').columns:
    print(f"{col}: {df[col].min()} - {df[col].max()}")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='species', data=df)
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=df)
plt.title('Flipper Length vs. Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['island'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Island Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='species', hue='sex', data=df)
plt.title('Sex Distribution within Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap of Correlations')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=df, ax=axes[0])
sns.boxplot(x='species', y='body_mass_g', data=df, ax=axes[1])
sns.boxplot(x='species', y='bill_depth_mm', data=df, ax=axes[2])
plt.suptitle('Boxplots for Measurements by Species')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.histplot(df['bill_length_mm'], kde=True, bins=20, ax=axes[0])
sns.histplot(df['bill_depth_mm'], kde=True, bins=20, ax=axes[1])
sns.histplot(df['flipper_length_mm'], kde=True, bins=20, ax=axes[2])
plt.suptitle('Histograms of Measurements')
plt.show()
```
