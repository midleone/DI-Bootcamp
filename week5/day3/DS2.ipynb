```python
from plotnine import ggplot, aes, geom_point

import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram for Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
pair_plot = sns.pairplot(iris, hue='species')
pair_plot.fig.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
import plotly.express as px

fig_3d_scatter = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                               color='species', symbol='species', opacity=0.7,
                               title='3D Scatter Plot of Sepal Length, Sepal Width, and Petal Length')
fig_3d_scatter.show()
```
