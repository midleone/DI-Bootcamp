```python
data = pd.read_csv('IRIS.csv')
```

```python
def classify_columns(data):
    column_categories = {}

    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            column_categories[column] = {'type': 'quantitative', 'reason': 'Numeric data.'}
        else:
            column_categories[column] = {'type': 'qualitative', 'reason': 'Non-numeric data.'}

    return column_categories

column_categories = classify_columns(data)

for column, info in column_categories.items():
    print(f"{column}: {info['type']} - {info['reason']}")
```
