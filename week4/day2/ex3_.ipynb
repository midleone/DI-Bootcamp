```python
from faker import Faker
fake = Faker()
random_data = []
for _ in range(100):
    name = fake.name()
    address = fake.address()
    email = fake.email()
    random_data.append({
        'name': name,
        'address': address,
        'email': email
    })

for data in random_data:
    print(f"Name: {data['name']}\nAddress: {data['address']}\nEmail: {data['email']}\n")
```

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(20)

ages = np.random.randint(20, 61, size=100)

income_levels = np.random.uniform(1, 100000, size=100)

data = pd.DataFrame({
    'Name': [fake.name() for _ in range(100)],
    'Address': [fake.address() for _ in range(100)],
    'Email': [fake.email() for _ in range(100)],
    'Age': ages,
    'Income': income_levels
})

print(data.head(10))
```
