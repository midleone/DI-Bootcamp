```python
import scipy.stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_value = scipy.mean(data)

median_value = scipy.median(data)

variance_value = scipy.var(data)

std_dev_value = scipy.std(data)

print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value:.2f}")
print(f"Standard Deviation: {std_dev_value:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean_value = 50
std_dev_value = 10

normal_distribution = norm(loc=mean_value, scale=std_dev_value)

x_values = np.linspace(mean_value - 4 * std_dev_value, mean_value + 4 * std_dev_value, 1000)

plt.plot(x_values, normal_distribution.pdf(x_values), label=f'Mean={mean_value}, Std Dev={std_dev_value}')
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind

np.random.seed(0)

data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=12, size=100)

t_statistic, p_value = ttest_ind(data1, data2)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("The difference between the two datasets is statistically significant. Reject the null hypothesis.")
else:
    print("There is not enough evidence to reject the null hypothesis.")
```

```python
import numpy as np
from scipy.stats import skew, kurtosis

data = np.random.normal(0, 1, 1000)

data_skewness = skew(data)
data_kurtosis = kurtosis(data)

print(f"Skewness of the dataset: {data_skewness:.4f}")
print(f"Kurtosis of the dataset: {data_kurtosis:.4f}")
```
